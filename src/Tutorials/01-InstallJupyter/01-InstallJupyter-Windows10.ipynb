# Tutorial:  Installing Jupyter (Windows 10)

<br>
<br>

⚠️ Many university computers already have Jupyter installed. If you are using a university computer, check whether Jupyter is installed by entering `jupyter` in the Windows search bar.

<br>
<br>


[Jupyter](https://jupyter.org) is a framework for creating interactive HTML pages that can be run in normal web browsers like Google Chrome and Firefox.  Jupyter is used extensively throughout this open textbook, and most of this textbook's contents were created using Jupyter.

The easiest way to install Jupyter is to install [Anaconda][https://anaconda.org].

Below are step-by-step instructions for (a) installing Anaconda, then (b) checking the Jupyter installation.

# Step 2:  &nbsp; Install Anaconda
[Back to Table of Contents](#toc)
<br>
<br>


1.  Go to [anaconda.com/distribution](https://www.anaconda.com/distribution/) , and select **Windows**

<br>
<br>

<img src="img/ss2-1.png" alt="screenshot" width=800/>

<br>
<br>

2.  Scroll down to the **Windows Installer** options, and download the **Python 3.7** version.


⚠️ Choose either **64 bit** or **32 bit**  &nbsp; &nbsp; (see [Step 1](#check-windows-version))



<br>
<br>

<img src="img/ss2-2.png" alt="screenshot1" width=800/>

<br>
<br>



3.  Open the installer and install the software.  Detailed installation instructions:  [(English)](https://docs.anaconda.com/anaconda/install/windows/)  [(日本語)](https://www.python.jp/install/anaconda/windows/install.html) [(日本語２)](https://weblabo.oscasierra.net/python-anaconda-install-windows/)

<a name="check-jupyter"></a>
# Step 3:  &nbsp; Check Jupyter installation
[Back to Table of Contents](#toc)

<br>
<br>


1. Enter `jupyter notebook` in the Windows search bar, then select the **Jupyter Notebook** app.

<br>
<br>

<img src="img/ss3-1.png" alt="screenshot" width=800/>

<br>
<br>



2. If you have multiple browsers, Windows may ask you which browser you want to use.  Choose any browser (e.g. Firefox, Google Chrome, Microsoft Edge, etc.)

<br>
<br>

<img src="img/ss3-2.png" alt="screenshot" width=800/>

<br>
<br>



3. When the Jupyter home page launches, select **New**, then select a **Python 3** Notebook.

<br>
<br>

<img src="img/ss3-3.png" alt="screenshot" width=800/>

<br>
<br>





4. When the notebook launches, enter `print( "Hello world!" )` in the cell, then select **Run**.

<br>
<br>

<img src="img/ss3-4.png" alt="screenshot" width=800/>

<br>
<br>





5. Verify that `Hello world!` is printed as output.

<br>
<br>

<img src="img/ss3-5.png" alt="screenshot" width=800/>

<br>
<br>
